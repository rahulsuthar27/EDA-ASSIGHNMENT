```python
#.Identify and remove outliers in the km_driven column using the IQR method
import numpy as np
Q1 = np.quantile(ab["km_driven"], 0.25)
Q3 = np.quantile(ab["km_driven"], 0.75)
IQR=Q3-Q1
Lower = Q1 - 1.5 * IQR
Upper = Q3 + 1.5 * IQR
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {Lower}")
print(f"Upper Bound: {Upper}")
ab_cleaned = ab[(ab["km_driven"] > Lower) & (ab["km_driven"] <Upper)]
print(f"Original Dataset Size: {ab.shape[0]}")
print(f"Cleaned Dataset Size: {ab_cleaned.shape[0]}")
remaining_outliers = ab_cleaned[(ab_cleaned["km_driven"] <= Lower) | (ab_cleaned["km_driven"] >= Upper)]
print(f"Remaining Outliers: {remaining_outliers.shape[0]}")
plt.figure(figsize=(8,5))
sns.boxplot(x=ab["km_driven"], color="red")
plt.title("Boxplot of km_driven (Before Removing Outliers)")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=ab_cleaned["km_driven"], color="green")
plt.title("Boxplot of km_driven (After Removing Outliers)")
plt.show()

```

```python
#13.Which bike names are priced significantly above the average price for their manufacturing year
yearly_stats = ab.groupby('year')['selling_price'].agg(['mean', 'std']).reset_index()

# Merge stats with original dataframe
ao= ab.merge(yearly_stats, on='year')

# Define the threshold (Mean + 1.5 * Standard Deviation)
ab['Threshold'] = ao["mean"]+1.5*ao["std"]

# Filter bikes priced significantly above the threshold
expensive_bikes = ab[ab['selling_price'] > ab['Threshold']][['name', 'year', 'selling_price']]

# Display result
print(expensive_bikes)
```

Car price is not significantly influenced by engine size alone.
Other factors such as brand, features, and mileage may play a more important role in determining the price.
Some cars with smaller engines may still be expensive due to luxury features and brand value.
A boxplot comparison between engine types showed no significant price difference between the two categories, reinforcing the weak relationship.

2️⃣ Relationship Between Annual Income and Car Price
It is often expected that higher-income customers purchase more expensive cars. However, our findings show that:

The correlation between annual income and car price was also weak (~0.1).
Customers from different income levels tend to buy cars in a similar price range, suggesting that income is not the sole determining factor.

When comparing average car prices across income brackets, there was no strong upward trend. This suggests that both middle and high-income individuals buy cars within a similar price range.

To determine whether engine size preference is linked to income level, we analyzed the average income of customers purchasing each engine type. The results indicated that:

No significant income difference was observed between the two engine types.

4️⃣ Key Insights from the Analysis
Engine size has a very weak influence on car price, suggesting that performance is not the primary pricing factor.
